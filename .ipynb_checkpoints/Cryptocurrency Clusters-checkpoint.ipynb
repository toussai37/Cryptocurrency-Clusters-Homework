```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
crypto_data_df = pd.read_csv(Path('Resources/crypto_data.csv'))
```

```python
crypto_data_df.head()
```

```python
is_trading_df = crypto_data_df[crypto_data_df["IsTrading"]==True]
is_trading_df.head()
```

```python
drop_istrading_df = is_trading_df.drop("IsTrading", axis=1)
drop_istrading_df.head()
```

```python
drop_nan_df = drop_istrading_df.dropna()
drop_nan_df.head()
```

```python
drop_coin_name_df = drop_nan_df.drop("CoinName", axis=1)
drop_coin_name_df.head()
```

```python
drop_unname_df = drop_coin_name_df.drop("Unnamed: 0", axis=1)
drop_unname_df.head()
```

```python
mined_df = drop_unname_df[drop_unname_df["TotalCoinsMined"] > 0]
mined_df.head()
```

```python
x_dummies = pd.get_dummies(mined_df)
x_dummies.head()
```

```python
x_scaled = StandardScaler().fit_transform(x_dummies)
```

```python
#377 features
```

```python
pca = PCA(n_components=0.90)
x_scaled_pca = pca.fit_transform(x_scaled)
x_scaled_pca_df = pd.DataFrame(data=x_scaled_pca)
x_scaled_pca_df.head()
```

#274 features

```python
#How did the number of the features change?
#the number of features droped by 103 when PCA was applied.
```

```python
tsne = TSNE(learning_rate = 35)
features = tsne.fit_transform(x_scaled_pca_df)
```

```python
x_scaled_pca_df['x'] = features[:,0]
x_scaled_pca_df['y'] = features[:,1]
plt.scatter(x_scaled_pca_df['x'],x_scaled_pca_df['y'])
plt.show()
```

```python
# Observe whether there are distinct clusters or not.
#there is an observable cluster.
```

```python
inertia = []
k = list(range(1,10))
for i in k:
    km = KMeans(n_clusters = i, random_state = 0)
    km.fit(x_scaled_pca_df)
    inertia.append(km.inertia_)

elbow_data = {'k': k, 'inertia': inertia}
elbow_data_df = pd.DataFrame(elbow_data)

```

```python
plt.plot(elbow_data_df['k'],elbow_data_df['inertia'])
plt.xticks(list(range(10)))
plt.title('Elbow Curve')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.show()        
```
